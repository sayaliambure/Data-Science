```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

## Apply OLS

```python
# Apply OLS
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
# m
#gives you an array of weights estimated by linear regression
```

```python
reg.intercept_
# Independent term in the linear model (b)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x), color='red')
# actual line drawn by model
```

## Apply gradient descent

* apply gradient descent assuming constant slope m=78.35
* assume starting random value of b is 0

```python
# y = mx + c
y_pred = ((78.35*x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color='red', label='OLS')  # line for OLS
plt.plot(x, y_pred, color='green', label='b=0')  #line of first iteration of gradient descent
plt.legend()
plt.show()
```

```python
m=78.35
b=0

# slope at b=0
# slope= -2 * summation(yi - mxi - b)
# .ravel() - function to make multi dimension array into 1D

loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

Step_size = learning rate * slope

```python
# Let learning rate = 0.1
lr = 0.1

step_size = loss_slope * lr
step_size
```

```python
# cal new b
b = b - step_size
b
```

```python
y_pred1 = ((78.35*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color='red', label='OLS')  # line for OLS
plt.plot(x, y_pred, color='green', label='b=0')  #line of first iteration of gradient descent
plt.plot(x, y_pred1, color='blue', label='b={}'.format(b))  #line of second iteration of gradient descent
plt.legend()
plt.show()
```

```python
# ext iteration
loss_slope = -2 * np.sum(y - m*x.ravel() - b)
print(loss_slope)

step_size = loss_slope * lr
print(step_size)

# cal new b
b = b - step_size
b
```

```python
y_pred2 = ((78.35*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color='red', label='OLS')  # line for OLS
plt.plot(x, y_pred, color='green', label='b=0')  #line of first iteration of gradient descent
plt.plot(x, y_pred1, color='blue', label='b={}'.format(b))  #line of second iteration of gradient descent
plt.plot(x, y_pred2, color='black', label='b={}'.format(b))  #line of second iteration of gradient descent

plt.legend()
plt.show()
```

```python
# ext iteration
loss_slope = -2 * np.sum(y - m*x.ravel() - b)
print(loss_slope)

step_size = loss_slope * lr
print(step_size)

# cal new b
b = b - step_size
b
```

```python
y_pred3 = ((78.35*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color='red', label='OLS')  # line for OLS
plt.plot(x, y_pred, color='green', label='b=0')  #line of first iteration of gradient descent
plt.plot(x, y_pred1, color='blue', label='b={}'.format(b))  #line of second iteration of gradient descent
plt.plot(x, y_pred2, color='black', label='b={}'.format(b))  #line of second iteration of gradient descent
plt.plot(x, y_pred3, color='orange', label='b={}'.format(b))  #line of second iteration of gradient descent

plt.legend()
plt.show()
```

```python
b = 400
m = 78.35
lr = 0.1 
epoch = 10

for i in range(epoch):
    loss_slope = -2 * np.sum(y - m*x.ravel() - b)
    b = b - (loss_slope * lr)
    y_pred = m*x + b
    
    plt.plot(x, y_pred)
    
plt.scatter(x, y)
    
```
